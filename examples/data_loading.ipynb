```python
from pyDrop.clustering.PCRData import PCRData
```

```python
# load and plot 3D data
filepath = "../data/3d_assay_6.csv"
data = PCRData(filepath, sep=",", header=0)
data.plot()
```

```python
# Do elbow and silhouette method to verify the number of samples (useful for 3D+ data)

data.predict_num_clusters(method="elbow", display=True)
```

```python
data.predict_num_clusters(method="silhouette", display=True)
```

```python
# subsample the data for testing
X, y = data.subsample_clusters({0:      0.0005,
                                1000:   0.5,
                                10000:  0.5,
                                11000:  0.5,
                                100000: 1,
                                101000: 1,
                                110000: 1,
                                111000: 1}, inplace=True)
data.plot()
```
